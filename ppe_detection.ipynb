# **REAL TIME CAPTURE**

Key	Function
C	Stops detection and webcam
S	Saves current frame as screenshot

```python
# ================= Setup =================
from IPython.display import display, Javascript, Image, clear_output
from google.colab.output import eval_js
from base64 import b64decode
import numpy as np
import cv2
import time
import os
import pandas as pd
from ultralytics import YOLO
from google.colab import output

# Load your model
model = YOLO('/content/runs/detect/ppe_yolo8s_model11/weights/best.pt')

# Setup directories
os.makedirs('/content/detections/screenshots', exist_ok=True)
log_file_path = '/content/detections/detection_log.csv'
if not os.path.exists(log_file_path):
    pd.DataFrame(columns=["timestamp", "class", "confidence"]).to_csv(log_file_path, index=False)

# State control
detection_active = True
stop_requested = False
frame_id = 0

# Required PPE
required_ppe = ["Helmet", "IFR Suit", "Safety Shoes"]

# ============== Callbacks ================
def stop_detection():
    global stop_requested
    stop_requested = True
    print("🟥 Detection stopped.")

output.register_callback('notebook.stopDetection', stop_detection)

def save_detection_log(results):
    log_data = []
    timestamp = time.strftime("%Y-%m-%d %H:%M:%S")
    for result in results:
        boxes = result.boxes
        for i in range(len(boxes.cls)):
            class_id = int(boxes.cls[i])
            conf = float(boxes.conf[i])
            name = model.names[class_id]
            log_data.append({"timestamp": timestamp, "class": name, "confidence": conf})
    if log_data:
        df = pd.DataFrame(log_data)
        df.to_csv(log_file_path, mode='a', index=False, header=False)

def handle_frame(data_url):
    global stop_requested, frame_id

    if stop_requested:
        return

    # Decode image
    binary = b64decode(data_url.split(',')[1])
    np_arr = np.frombuffer(binary, np.uint8)
    img = cv2.imdecode(np_arr, cv2.IMREAD_COLOR)

    # Run detection
    results = model.predict(source=img, conf=0.4, verbose=False)
    detected_classes = set()

    for result in results:
        img = result.plot()
        detected_classes.update([model.names[int(cls)] for cls in result.boxes.cls])

    save_detection_log(results)

    # Check for missing PPE
    missing_ppe = [item for item in required_ppe if item not in detected_classes]
    if missing_ppe:
        cv2.putText(img, f"Missing: {', '.join(missing_ppe)}", (10, 30),
                    cv2.FONT_HERSHEY_SIMPLEX, 0.8, (0, 0, 255), 2)

    # Screenshot if requested
    if 'screenshot' in data_url:
        filename = f"/content/detections/screenshots/frame_{frame_id}.jpg"
        cv2.imwrite(filename, img)
        print(f"📸 Screenshot saved: {filename}")
        frame_id += 1

    # Show image
    _, im_buf_arr = cv2.imencode('.jpg', img)
    clear_output(wait=True)
    display(Image(data=im_buf_arr.tobytes()))
    print("🟢 Detecting PPE... (Press 'S' to save screenshot, 'C' to stop)")

output.register_callback('notebook.handle_frame', handle_frame)

# ============== JS Webcam Code ==============
js = Javascript("""
let video, canvas, ctx, stream;
let stopped = false;

function stopCam() {
    if (stream) {
        stream.getTracks().forEach(track => track.stop());
    }
    stopped = true;
}

async function startCamera() {
    video = document.createElement('video');
    video.setAttribute('playsinline', '');
    video.style = 'width: 100%; max-width: 480px; display: block;';
    document.body.appendChild(video);

    canvas = document.createElement('canvas');
    ctx = canvas.getContext('2d');
    document.body.appendChild(canvas);

    try {
        stream = await navigator.mediaDevices.getUserMedia({ video: true });
        video.srcObject = stream;
        await video.play();
        requestAnimationFrame(captureFrame);
    } catch (e) {
        alert('Error accessing webcam.');
    }

    document.addEventListener('keydown', function(event) {
        if (event.code === 'KeyC') {
            stopCam();
            google.colab.kernel.invokeFunction('notebook.stopDetection', [], {});
        } else if (event.code === 'KeyS') {
            captureFrame(true);  // force screenshot
        }
    });
}

function captureFrame(screenshot = false) {
    if (stopped || !video.videoWidth) return;

    canvas.width = video.videoWidth;
    canvas.height = video.videoHeight;
    ctx.drawImage(video, 0, 0);

    let dataUrl = canvas.toDataURL('image/jpeg', 0.8);
    if (screenshot) dataUrl += "screenshot";  // flag screenshot
    google.colab.kernel.invokeFunction('notebook.handle_frame', [dataUrl], {});
    requestAnimationFrame(captureFrame);
}

startCamera();
""")

display(js)
```
